```python
data_csv['date'] = pd.to_datetime(data_csv['date'],format='%Y-%m-%d')

y_data = data_csv.loc[(data_csv['date'] >= '2022-01-03')
                     & (data_csv['date'] < '2022-04-29')]
```

```python
monthvise= data_csv.groupby(data_csv['date'].dt.strftime('%B'))[['open','close']].mean()
new_order = ['January', 'February', 'March', 'April']
monthvise = monthvise.reindex(new_order, axis=0)
monthvise
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Bar(x=monthvise.index, y=monthvise['open'], name='open'))
fig.add_trace(go.Bar(x=monthvise.index, y=monthvise['close'], name='close'))
fig.update_layout(barmode='group')
fig.show()
```

```python
monthvise2= data_csv.groupby(data_csv['date'].dt.strftime('%B'))[['high','low']].mean()
new_order = ['January', 'February', 'March', 'April']
monthvise2 = monthvise2.reindex(new_order, axis=0)
monthvise2
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=monthvise2.index, y=monthvise2['high'], name='high'))
fig.add_trace(go.Bar(x=monthvise2.index, y=monthvise2['low'], name='low'))
fig.update_layout(barmode='group')
fig.show()
```
